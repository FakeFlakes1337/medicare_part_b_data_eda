```python
grouped_df = df.groupby('Rndrng_Prvdr_Org_Name').agg({
    'Avg_Tot_Sbmtd_Chrgs': 'sum',
    'Avg_Mdcr_Alowd_Amt': 'sum',
    'Avg_Mdcr_Pymt_Amt': 'sum',
}).round(1)
grouped_df
```

```python
# APC_Desc is a list aggregation of multiple levels
# Hence,one row can have multiple levels, so we are just estimating by retrieving the first listed level
def extract_APC_Desc_level(row):
    
    if 'Level' in row['APC_Desc']:
        return row['APC_Desc'][:8]
    return 'unknown'

df['APC_Desc_level'] = df.apply(extract_APC_Desc_level, axis=1)
```

```python
df
```

```python
# Level 3,2 and 4 are the top 3 common level claims
df['APC_Desc_level'].value_counts()
```

```python
### Attempting to view the distribution as a bar graph
value_counts_df = df['APC_Desc_level'].value_counts().reset_index()
value_counts_df.columns = ['APC_Desc_level', 'Count']

value_counts_df['percentage'] = (value_counts_df['Count']*100/ value_counts_df['Count'].sum()).round(1)


# Creating the bar plot using Seaborn
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
sns.barplot(data=value_counts_df, x='APC_Desc_level', y='percentage')
plt.title('Count of APC_Desc_level')
plt.xlabel('APC_Desc_level')
plt.ylabel('percentage')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```
