```r
library(readxl)
df <- read_excel("Y2020_0001U.xlsx")
```

```r
nrow(df)
```

```r
summary(df)
```

```r
df <- df%>%
    mutate(MODIFIER = if_else(is.na(MODIFIER), "NA", MODIFIER))
```

```r
df2 <- df %>%
     mutate(DESCRIPTION = as.factor(DESCRIPTION), 
           HCPCS = as.factor(HCPCS),
           MODIFIER       = as.factor(MODIFIER))
df2
```

```r
summary(df2)
```

```r
df3 <- df2 %>%
    filter(MODIFIER =='TOTAL') %>%
    arrange(desc(PAYMENT))
df3
```

```r
df4 <- df2 %>%
  group_by(MODIFIER) %>%
  summarize(MEDIAN_PAYMENT = median(PAYMENT)) %>%
  ungroup() %>%
  arrange(desc(MEDIAN_PAYMENT)) 
df4
```
